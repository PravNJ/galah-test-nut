```python
from specutils import Spectrum1D
from specutils import SpectralRegion
from specutils.analysis import equivalent_width
from specutils.analysis import fwhm
```

```python
import h5py
hf_grid = h5py.File("/data/praveen/wl_grid.h5","r")
wl_grid = np.array(hf_grid["wl_grid"])
```

```python
grid = wl_grid[PLOT_MASK]
```

```python
df_difference_masked["EW"] = df_difference_masked.apply(lambda row : equivalent_width(Spectrum1D(spectral_axis=grid*u.AA,flux=row.to_numpy()*u.Unit('erg cm-2 s-1 AA-1'))).value, axis=1)
```
